```python
import openai

# Set your OpenAI API key
openai.api_key = 'YOUR_OPENAI_API_KEY'

def predict_sentiment(fine_tuned_model, new_paragraph):
    prompt = "<start>\nParagraph: " + new_paragraph.strip() + "\nPrediction:"
    
    response = openai.Completion.create(
        model=fine_tuned_model,
        prompt=prompt,
        max_tokens=10,
        temperature=0,
        stop="<end>"
    )
    
    # Extract the prediction
    prediction = response.choices[0].text.strip()
    return prediction

# Example usage
if __name__ == "__main__":
    fine_tuned_model = 'your-fine-tuned-model-name'  # Replace with your fine-tuned model's name
    new_paragraph = "The book had some interesting ideas, but overall it failed to capture my attention."
    
    prediction = predict_sentiment(fine_tuned_model, new_paragraph)
    print(f"Prediction:\n{prediction}")
```
